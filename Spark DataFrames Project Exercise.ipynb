#### Start a simple Spark Session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

#### Load the Walmart Stock CSV File, have Spark infer the data types.

```python
# File location and type
file_location = "/FileStore/tables/walmart_stock.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
```

#### What are the column names?

```python
df.columns
```

#### What does the Schema look like?

```python
df.printSchema()
```

#### Print out the first 5 columns.

```python
for i in range(5):
    print(df.select("Date", "Open", "High", "Low", "Close").collect()[i])
```

#### Use describe() to learn about the DataFrame.

```python
df.describe().show()
```

#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. #### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
df2 = df.withColumn('HV Ratio', df["High"]/df["Volume"])
```

```python
df2.show()
```

#### What day had the Peak High in Price?

```python
df2.agg({"High": "max"}).collect()[0]
```

#### What is the mean of the Close column?

```python
df2.agg({"Close": "avg"}).collect()[0]
```

#### What is the max and min of the Volume column?

```python
df2.agg({"Volume": "max"}).collect()[0]
```

```python
df2.agg({"Volume": "min"}).collect()[0]
```

#### How many days was the Close lower than 60 dollars?

```python
df2[df2["Close"] < 60].count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
df2[df2["High"] > 80].count()/df.count()
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```python
df2.corr("High","Volume")
```

#### What is the max High per year?

```python
df2.agg({"High": "max"}).collect()[0]
```

```python
from pyspark.sql.functions import year, max

df2.groupBy(year("Date")).agg(max("High")).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```python
from pyspark.sql.functions import month, mean

df2.groupBy(month("Date")).agg(mean("High")).show()
```
